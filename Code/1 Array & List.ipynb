# Given a list of heterogenous elements. Write a python script to remove all the non int values from the list.

```python
lst = [1, 'apple', 3.14, 'banana', 5, 'orange', 'grape', 7]
# Initializing an empty list to store integer values
int_list = []

# Loop through each element in the original list
for x in lst:
    # Check if the element is an integer
    if isinstance(x, int):
        # Append the integer value to the new list
        int_list.append(x)

print("Original list:", lst)
print("List with only integers:", int_list)
```

# Write a Python Script to create a list of first N prime numbers.

```python
def is_prime(num):
    """Check if a number is prime."""
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def generate_primes(n):
    """Generate a list of the first N prime numbers."""
    primes = []
    num = 2  # Start checking from the first prime number
    while len(primes) < n:
        if is_prime(num):
            primes.append(num)
        num += 1
    return primes

# Specify the value of N
N = 10  # You can change this to any desired number

# Generate and display the list of first N prime numbers
prime_list = generate_primes(N)
print(f"The first {N} prime numbers are:", prime_list)
```


# Write a Python Script to create a list of first N terms of a Fibonacci series.

```python
def generate_fibonacci(n):
    """Generate a list of the first N terms of a Fibonacci series."""
    fibonacci_list = []
    a, b = 0, 1  # Initialize the first two terms of the Fibonacci series

    for _ in range(n):
        fibonacci_list.append(a)
        a, b = b, a + b  # Update terms for the next iteration

    return fibonacci_list

# Specify the value of N
N = 10  # You can change this to any desired number

# Generate and display the list of first N terms of the Fibonacci series
fibonacci_series = generate_fibonacci(N)
print(f"The first {N} terms of the Fibonacci series are:", fibonacci_series)
```
