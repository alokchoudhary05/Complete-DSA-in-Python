# **Priority Queue**

## **Priority Queue Using List**

Assignment - Priority queue using list.jpg

```python
""" Priority Queue Using List """
# 1.
class PriorityQueue:
  def __init__(self):
    self.items = []
# 2.
  def push(self,data,priority):
    index = 0
    while index<len(self.items) and self.items[index][1]<=priority:
      index += 1
    self.items.insert(index,(data,priority))
# 4.
  def is_empty(self):
    return len(self.items)==0
# 3.
  def pop(self):
    if self.is_empty():
      raise IndexError("Priority Queue is empty")
    return self.items.pop(0)[0]
# 5.
  def size(self):
    return len(self.items)

# Driver Code
p = PriorityQueue()
p.push("Alok",1)
p.push("Ayush",5)
p.push("Shub",4)
p.push("Rajeev",7)
p.push("Aman",8)
p.push("Omjee",3)

while not p.is_empty():
  print(p.pop())
```

## **Priority Queue Using Linked List**

Assignment - Priority queue using linked list.jpg

```python
""" Priority Queue Using Linked List """
# 1.
class Node:
  def __init__(self,item=None,priority=None,next=None):
    self.item = item
    self.priority = priority
    self.next = next
# 2.
class PriorityQueue:
  def __init__(self):
    self.start = None
    self.item_count = 0
# 3.
  def push(self,data,priority):
    n = Node(data,priority)
    if not self.start or priority < self.start.priority:
      n.next = self.start
      self.start = n
    else:
      temp = self.start
      while temp.next and temp.next.priority <= priority:
        temp = temp.next
      n.next = temp.next
      temp.next = n
    self.item_count += 1
# 5.
  def is_empty(self):
    return self.start == None
# 4.
  def pop(self):
    if self.is_empty():
      raise IndexError("Priority Queue is empty")
    data = self.start.item
    self.start = self.start.next
    self.item_count -= 1
    return data
# 6.
  def size(self):
    return self.item_count

# Driver Code
p = PriorityQueue()
p.push("Alok",1)
p.push("Ayush",5)
p.push("Shub",4)
p.push("Rajeev",7)
p.push("Aman",8)
p.push("Omjee",3)

while not p.is_empty():
  print(p.pop())
```
