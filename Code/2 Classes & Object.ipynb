# Create a class Employee with attributes empid, name, salary and also define methods to access properties of Employee.

```python
class Employee:
  def __init__(self,empid=None,name=None,salary=None):
    self.empid=empid
    self.name=name
    self.salary=salary

  def setEmpid(self,empid):
    self.empid=empid
  def setName(self,name):
    self.name=name
  def setSalary(self,salary):
    self.salary=salary

  def getEmpid(self):
    return self.empid
  def getName(self):
    return self.name
  def getSalary(self):
    return self.salary

e1 = Employee()
e2 = Employee(101,"Aman",50000)
e1.setEmpid(102)
e1.setName("John")
e1.setSalary(20000)
e3 = Employee(103,"Khonw",60000)

print(e1.getEmpid(),e1.getName(),e1.getSalary())
print(e2.getEmpid(),e2.getName(),e2.getSalary())
print(e3.getEmpid(),e3.getName(),e3.getSalary())
```

# Define a Python class Person with instance object variables name and age. Set instance object variable in  __ init__() method. Also define show() method to disply name and age of a person.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(f"Name: {self.name}, Age: {self.age}")

p1 = Person("John", 25)
p2 = Person("Doner",19)
p3 = Person("Forera",22)
p1.show()
p2.show()
p3.show()
```

# Define a class circle with instance object variable radius.provide setter and getter for radius. Also define getArea() and getCircunference() method.

```python
import math
class Circle:
  def __init__(self,radius):
    self.radius = radius

  def setRadius(self,radius):
    self.radius = radius

  def getArea(self):
    return math.pi*self.radius**2
  def getCircumference(self):
    return 2*math.pi*self.radius

c1=Circle(58)
print(f"Area = ",c1.getArea())
print("Circumference = ",c1.getCircumference())
```

```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def setDimensions(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def showDimensions(self):
        print("Length:", self.length)
        print("Breadth:", self.breadth)

    def getArea(self):
        return self.length * self.breadth

# Example usage:
rectangle = Rectangle(5, 4)
rectangle.showDimensions()
print("Area:", rectangle.getArea())

# Modify dimensions
rectangle.setDimensions(7, 3)
rectangle.showDimensions()
print("Area:", rectangle.getArea())
```

```python
class Team:
  def __init__(self):
    self.members = []

  def inputMembers(self):
    num_members = int(input("Enter the number of group member: "))
    for i in range(num_members):
            member_name = input(f"Enter name of team member {i+1}: ")
            self.members.append(member_name)

  def displayMembers(self):
        print("Team Members:")
        for member in self.members:
            print(member)

team = Team()
team.inputMembers()
team.displayMembers()
```
