## **Deque Implementation Using List**

image.png

```python
""" Deque Implementation Using List """
# 1.
class Deque:
  def __init__(self):
    self.items = []
# 2.
  def is_empty(self):
    return len(self.items)==0
# 3.
  def insert_front(self,data):
    self.items.insert(0,data)
# 4.
  def insert_rear(self,data):
    self.items.append(data)
# 5.
  def delete_front(self):
    if self.is_empty():
      raise IndexError("Deque is Empty")
    else:
      self.items.pop()
# 6.
  def delete_rear(self):
    if self.is_empty():
      raise IndexError("Deque is Empty")
    else:
      self.items.pop()
# 7.
  def get_front(self):
    if self.is_empty():
      raise IndexError("Deque is Empty")
    else:
      return self.items[0]
# 8.
  def get_rear(self):
    if self.is_empty():
      raise IndexError("Deque is Empty")
    else:
      return self.items[-1]
# 9.
  def size(self):
    return len(self.items)

# Driver Code
d1 = Deque()
d1.insert_front(10)
d1.insert_front(20)
d1.insert_rear(30)
d1.insert_rear(40)
print("Front =",d1.get_front(),"Rear =",d1.get_rear())
```

## **Deque Using Doubly Linked List**

```python
""" Deque Using Doubly Linked List """
# 1.
class Node:
  def __init__(self,item=None,prev=None,next=None):
    self.prev = prev
    self.item = item
    self.next = next
# 2.
class Deque:
  def __init__(self):
    self.front = None
    self.rear = None
    self.item_count = 0
# 3.
  def is_empty(self):
    return self.item_count == 0
# 4.
  def insert_front(self,data):
    n=Node(data,None,self.front)
    if self.is_empty():
      self.rear = n
    else:
      self.front.prev = n
    self.front = n
    self.item_count += 1
# 5.
  def insert_rear(self,data):
    n = Node(data,self.rear)
    if self.is_empty():
      self.rear = n
    else:
      self.rear.next = n
    self.rear = n
    self.item_count += 1
# 6.
  def delete_front(self):
    if self.is_empty():
      raise IndexError("Deque is empty")
    if self.front == self.rear:
      self.front = None
      self.rear = None
    else:
      self.front = self.front.next
      self.front.prev = None
    self.item_count -= 1
# 7.
  def delete_rear(self):
    if self.is_empty():
      raise IndexError("Deque is empty")
    if self.front == self.rear:
      self.front = None
      self.rear = None
    else:
      self.rear = self.rear.prev
      self.rear = None
    self.item_count -= 1
# 8.
  def get_front(self):
    if self.is_empty():
      raise IndexError("Deque is empty")
    return self.front.item
# 9.
  def get_rear(self):
    if self.is_empty():
      raise IndexError("Deque is empty")
    return self.rear.item
# 10.
  def size(self):
    return self.item_count

# Driver Code
d1 = Deque()
d1.insert_front(10)
d1.insert_front(20)
d1.insert_rear(30)
d1.insert_rear(40)
print("Front =",d1.get_front(),"Rear =",d1.get_rear())
# print("Delete Front =",d1.delete_front(),"Delete Rear =",d1.delete_rear())
print("Item = ",d1.item_count)
```
