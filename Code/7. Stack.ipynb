# **Stack Using List**

image.png

```python
""" Stack Using List """
# 1.
class Stack:
  def __init__(self):
    self.items = []
# 2.
  def is_empty(self):
    return len(self.items) == 0
# 3.
  def push(self,data):
    self.items.append(data)
# 4.
  def pop(self):
    if not self.is_empty():
      return self.items.pop()
    else:
      raise IndexError("Stack is Empty")
# 5.
  def peek(self):
    if not self.is_empty():
      return self.items[-1]
    else:
      raise IndexError("Stack is Empty")
# 6.
  def size(self):
    return len(self.items)

# Driver Code
s1 = Stack()
s1.push(10)
s1.push(20)
s1.push(30)
print("Top element is",s1.peek())
print("Removed element",s1.pop())
print("Top element is",s1.peek())
print()
```

# **Stack Using Linked List Concept**

image.png

```python
""" Stack Using Linked List Concept """
# 1.
class Node:
  def __init__(self,item=None,next=None):
    self.item = item
    self.next = next
class Stack:
  def __init__(self):
    self.start = None
    self.item_count = 0
# 2.
  def is_empty(self):
    return self.start == None
# 3.
  def push(self,data):
    n = Node(data,self.start)
    self.start = n
    self.item_count += 1
# 4.
  def pop(self):
    if not self.is_empty():
      data = self.start.item
      self.start = self.start.next
      self.item_count -= 1
      return data
    else:
      raise IndexError("Stack is Empty")
# 5.
  def peek(self):
    if not self.is_empty():
      return self.start.item
    else:
      raise IndexError("Stack is Empty")
# 6.
  def size(self):
    return self.item_count

# Driver Code
s1 = Stack()
s1.push(10)
s1.push(20)
s1.push(30)
s1.push(40)
print("Total element in the Stack=",s1.size())
print("Top element on the Stack is",s1.peek())
print("Delete Element is",s1.pop())
print("Total element in the Stack=",s1.size())
print("Top element on the Stack is",s1.peek())
```
